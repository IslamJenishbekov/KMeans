```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
coords = np.concatenate([np.random.randn(100, 3)+np.array([1, 2, -1]),
                         np.random.randn(100, 3)+np.array([-5, 4, 2]),
                         np.random.randn(150, 3)+np.array([3, -3, 0]),
                         np.random.randn(150, 3)+np.array([7, 1, -10])], axis=0)

fig = go.Figure(data=[go.Scatter3d(
    x=coords[:, 0],
    y=coords[:, 1],
    z=coords[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color='red'
    )
)])

fig.show()
```

```python
def init_centroids(num_clasters:int, data:np.ndarray)->np.ndarray:
  x = np.random.randint(data[:, 0].min(), data[:, 0].max(), (num_clasters, 1))
  y = np.random.randint(data[:, 1].min(), data[:, 1].max(), (num_clasters, 1))
  z = np.random.randint(data[:, 2].min(), data[:, 2].max(), (num_clasters, 1))
  return np.array([x, y, z]).T[0]

def centr_dist(data:np.ndarray, centroid:np.ndarray)->np.ndarray:
  return np.linalg.norm(data - centroid, axis=1)
```

```python
centroids = init_centroids(4, coords)
```

```python


fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=coords[:, 0],
    y=coords[:, 1],
    z=coords[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color='red'  # Цвет первого набора данных
    ),
    name='Dataset 1'

))

fig.add_trace(go.Scatter3d(
    x=centroids[:, 0],
    y=centroids[:, 1],
    z=centroids[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color='blue'  # Цвет первого набора данных
    ),
    name='Dataset 2'

))

fig.show()
```

```python
N_CLASTERS = 4
N_ITER = 10

for _ in range(N_ITER):
  labels = np.argmin(np.array([centr_dist(coords, centroids[i]) for i in range(N_CLASTERS)]).T, axis=1)
  centroids = np.array([coords[labels==i].mean(axis=0) for i in range(N_CLASTERS)])

labels = np.argmin(np.array([centr_dist(coords, centroids[i]) for i in range(N_CLASTERS)]).T, axis=1)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=coords[:, 0],
    y=coords[:, 1],
    z=coords[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color=labels,  # Цвет первого набора данных
        colorscale='Viridis',  # Выбираем цветовую схему
        colorbar=dict(title='Цветовая шкала')
    ),
    name='coords'

))

fig.add_trace(go.Scatter3d(
    x=centroids[:, 0],
    y=centroids[:, 1],
    z=centroids[:, 2],
    mode='markers',
    marker=dict(
        size=10,
        color='red'  # Цвет первого набора данных
    ),
    name='centroids'

))

fig.show()
```
